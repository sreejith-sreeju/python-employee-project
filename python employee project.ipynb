```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("EmpData.csv")
data.info()
```

```python
data["Height"] = np.random.randint(150, 181, size=len(data))
```

### How many are there in each Team and the percentage splitting with respect to the total employees.

```python
Team_count= data["Team"].value_counts()
Team_percentage= Team_count/len(data)*100

print("Team count: ")
print(Team_count)
print("\nTeam percentage: ")
print(Team_percentage)
```

## Segregate the employees w.r.t different positions


```python
position_count= data["Position"].value_counts()

print("position count: ")
print(position_count)
```

## Age group to which most employees belong

```python
age_group= pd.cut(data["Age"], bins= [18,25,35,45,55,65],labels= ["18-25","26-35","36-45","46-55","56-65"])
common_age = age_group.mode()[0]

print("Most_age_group_are_between: ",common_age)
```

##  Team and position the spending in terms of salary is high

```python
highest_salary = data.groupby(["Team","Position"])["Salary"].sum().nlargest(1).index[0]

print("Highest salary team and their position",highest_salary)
```

##  correlation between age and salary, and represent it visually

```python
plt.scatter(data["Age"], data["Salary"])
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Correlation between age and salary")
plt.show()
```
